```python
import numpy as np
from matplotlib.projections.geo import MollweideAxes
import matplotlib.projections as mprojections
from matplotlib.axes import Axes
from matplotlib.patches import Wedge
import matplotlib.spines as mspines

class LowerMollweideAxes(MollweideAxes):
    name = 'lower_hammer'
    def cla(self):
        MollweideAxes.cla(self)
        Axes.set_xlim(self, 0, np.pi / 2)
        Axes.set_ylim(self, np.pi / 2.0, 0)

    def _gen_axes_patch(self):
        # return Wedge((0.5, 0.5), 0.5, 180, 360)
        return Wedge((0.5, 0.5), 0.5, 0, 90)

    def _gen_axes_spines(self):
        # path = Wedge((0, 0), 1.0, 180, 360).get_path()
        path = Wedge((0, 0), 1.0, 0, 90).get_path()
        spine = mspines.Spine(self, 'circle', path)
        spine.set_patch_circle((0.5, 0.5), 0.5)
        return {'wedge':spine}

mprojections.register_projection(LowerMollweideAxes)

if __name__ == '__main__':
    import matplotlib.pyplot as plt
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='lower_hammer')
    ax.grid(True)
    plt.show()
```
